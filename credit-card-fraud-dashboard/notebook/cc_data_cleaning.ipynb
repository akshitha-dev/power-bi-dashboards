```python
import pandas as pd
import numpy as np
```

```python
credit=pd.read_csv("C:\\Users\\akshi\\OneDrive\\Desktop\\fraudTest_uncleaned.csv")
```

```python
credit.head()
```

```python
credit['trans_date_trans_time']=pd.to_datetime(credit['trans_date_trans_time'],format='%Y/%m/%d')
```

```python
credit.dtypes
```

```python
credit.isnull().sum()
```

```python
credit['amt']=pd.to_numeric(credit['amt'],errors='coerce')
credit['amt'].fillna(credit['amt'].median())
```

```python
credit['category']=credit['category'].fillna('unknown')
credit['merchant']=credit['merchant'].fillna('unknown')
credit['job']=credit['job'].fillna('Not provided')
credit['first']=credit['first'].fillna('Missing')
credit['city']=credit['city'].fillna('unknown')
```

```python
credit['trans_date_trans_time']=credit['trans_date_trans_time'].fillna('ffill')
```

```python
credit.rename(columns={
    'trans_date_trans_time': 'transaction_date',
    'amt': 'amount_usd',
    'category': 'purchase_category',
'':'id'})
```

```python
credit.duplicated().sum()
```

```python
credit.isnull().sum()
```

```python
credit['first'] =credit['first'].astype(str).str.replace(r'[^\w\s]', '', regex=True).str.strip().str.title()
credit['last'] =  credit['last'].astype(str).str.replace(r'[^\w\s]', '', regex=True).str.strip().str.title()
```

```python
credit.to_csv(r"C:\Users\akshi\OneDrive\Desktop\fraudtest_cleaned.csv",index=False)
```
